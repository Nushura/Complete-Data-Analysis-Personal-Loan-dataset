# Correlations -- Using Seaborn

The greater the strength of the correlation between attributes, the more close is its correlation value to 1 or the darker is its color. And hence, the closer the correlation value to 0 or the lighter the color is means that the strength strength between relationships are weaker. 

That is represented by the code and result below. Correlation is computed using seaborn's heatmap function.

```python
# Using seaborn library to plot the correlations as a heatmap
# 20, 20 is the width and hight of the heat map
plt.rcParams['figure.figsize'] = (20, 20)
sns.heatmap(personal_loan_df.corr(), annot = True, linewidths=.5, cmap="YlGnBu")

# Title of the heat map
plt.title('Correlation between features of\nPersonal Loan Dataset')

# Saving the graph to disk

# automatically adjusts subplot to fit in the figure area
plt.tight_layout()

# dpi=300 is print quality
plt.savefig('correlation-heatmap-rectangular.png', dpi=300) 
```

It can be observed that the heatmap removed a few attributes that contained strings. The personal_loan outcome column is hence removed as well from the heatmap. 

However, by observing the heatmap, it can be concluded that the following attributes have the strongest correlation with each other, which also contributes to determining if an individual will take the loan. The attributes and their correlation values are:

1. Income and mortgage_amt: 0.13
2. Income and credit_card_spend: 0.15
3. Income and fixed_deposit_acct: 0.19
4. mortgage_amt and fixed_deposit_acct: 0.1

Hence, income, mortgage_amt, credit_card_spend, and fixed_deposit_acct play a vital role in determining the probability of a particular individual taking a personal loan. For example, if an individual earns above a certain amount, they are most likely to take a loan as clients consider their chances of paying back before taking loans. 

Credit card spending will also increase the chances of someone taking a loan. This is because if their monthly expenditure is above a specific limit, then they will not lean towards taking any loans. People with fixed deposit accounts are likely to save money and hence can take a loan.

### Impact of credit card spend (Discrete variable) on loans

```python
# Categorising the fixed_deposit_acct column to have yes and no categories instead of 1 and 0
def credit_card_spend_category(df): 
    if (df["credit_card_spend"] == 1) & (df["credit_card_spend"] <= 1.5):
        return "normal_sepnd"
    elif df["credit_card_spend"] > 1.5:
        return "high_spend"

    
# Use apply() to apply a lambda function for all the rows of the dataframe
# The lambda function calls the previously defined function fixed_deposit_acct_category()
# To which it passes the personal_loan_df as parameter
# And a category is returned and saved in a new column named "fixed_deposit_acct_category"
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html
# the backslash is a line continuation character - there should be nothing after it, not even a space
personal_loan_df["credit_card_spend_category"] = \
    personal_loan_df.apply(lambda personal_loan_df:credit_card_spend_category(personal_loan_df), axis = 1) # axis=1 = row-wise
# Checking if implementation is proper
personal_loan_df.head()
```

```python
# Using seaborn to plot charts

# Plotting two separate plots side by side
# As a single plot
sns.countplot(x='credit_card_spend_category', 
              data=personal_loan_df)

plt.title('Credit Card Spend - Loaned vs Not Loaned', fontsize=30)
plt.xlabel("Credit Card Spend", fontsize=30)
plt.ylabel("Count", fontsize=30)
plt.tick_params(axis='both', which='both', labelsize=20) # x/y axes, minor/major ticks for plt

# Save the graph to disk
# plt.tight_layout() # automatically adjusts subplot to fit in the figure area
plt.savefig('Credit Card Spend.png', dpi=300)
```

### Impact of fixed deposit account (Categorical variable) on loans

```python
# Categorising the fixed_deposit_acct column to have yes and no categories instead of 1 and 0
def fixed_deposit_acct_category(df): 
    if df["fixed_deposit_acct"] == 1:
        return "Fixed Deposit Account"
    elif df["fixed_deposit_acct"] == 0:
        return "No Fixed Deposit Account"

    
# Use apply() to apply a lambda function for all the rows of the dataframe
# The lambda function calls the previously defined function fixed_deposit_acct_category()
# To which it passes the personal_loan_df as parameter
# And a category is returned and saved in a new column named "fixed_deposit_acct_category"
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html
# the backslash is a line continuation character - there should be nothing after it, not even a space
personal_loan_df["fixed_deposit_acct_category"] = \
    personal_loan_df.apply(lambda personal_loan_df:fixed_deposit_acct_category(personal_loan_df), axis = 1) # axis=1 = row-wise
# Checking if implementation is proper
personal_loan_df.head()
```

```python
# Using seaborn to plot charts

# Plotting two separate plots side by side
# As a single plot
sns.countplot(x='fixed_deposit_acct_category', 
              data=personal_loan_df, # Whole dataset 
              hue='fixed_deposit_acct') # grouped by outcome

plt.title('Fixed Deposit Account - Loaned vs Not Loaned', fontsize=30)
plt.xlabel("Fixed Deposit Account", fontsize=30)
plt.ylabel("Count", fontsize=30)
plt.tick_params(axis='both', which='both', labelsize=20) # x/y axes, minor/major ticks for plt


# If you want to save the graph to disk
plt.tight_layout() # automatically1m adjusts subplot to fit in the figure area
plt.savefig('fixed_deposit_acct_category-side-by-side.png', dpi=300) # dpi=300 is print quality
```
